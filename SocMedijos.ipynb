```python
import pandas as pd
import re

df = pd.read_csv('data.csv')
names = pd.read_csv('names.csv')

n_df = df[(df['CD_MemberList_Size'] > 10) & (df['CD_MemberList_Size'] < 30)].sort_values(by=['CD_MemberList_Size']) 

```

```python
for index, row in n_df.iterrows():
    print("");
    print("--------------------NEW GROUP----------------------");
    print("");
    words = re.split("\s+", row.at['CD_MemberList'])
    for item in words:
        name = names.loc[names['name'] == int(item)]['label']
        print(item + " ---> " +name.values[0]);
        
```
